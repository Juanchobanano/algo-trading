```python
# Import libraries.
import pandas as pd
```

```python
# Set magic num.
magic_num = str(4321)
```

```python
columns_order_data = ['Open Time', 'Order', 'Symbol', 'Type', 'Volume', 'Price', 'S / L', 'T / P', 'Time', 'State', 'Comment']
columns_deal_data = "date_time deal symbol type direction volume price order commission swap profit balance comment random".split()

# Get Order Data.
order_data = pd.read_csv("expert_" + magic_num +"_order_data.csv")
order_data = order_data[columns_order_data]

# Get Deal Data.
deal_data = pd.read_csv("expert_" + magic_num + "_deals_data.csv")
deal_data.columns = columns_deal_data
deal_data.drop("random", inplace=True, axis = 1)
```

```python
order_data.tail()
```

```python
deal_data.tail()
```

```python
deal_data["take_profit"] = order_data['T / P']
deal_data["stop_loss"] = order_data['S / L']
```

```python
columns = "date_time deal symbol type direction volume price take_profit stop_loss order commission swap profit balance comment".split()
new_columns = "date_time deal symbol order_type direction volume price take_profit stop_loss order_num commission swap profit balance comment".split()

# Generate new pandas dataframe.
deal_data = deal_data[columns]
deal_data.columns = new_columns
deal_data = deal_data[1:]
deal_data.head(10)
```

```python
# Format columns.
deal_data['date_time'] = pd.to_datetime(deal_data['date_time'])
deal_data["balance"] = deal_data.balance.map(lambda balance: float(balance.replace(" ", "")))
deal_data["deal"] = deal_data.deal.map(lambda deal: int(deal))
deal_data["volume"] = deal_data.volume.map(lambda volume: float(volume))
deal_data["price"] = deal_data.price.map(lambda price: float(price))
deal_data["take_profit"] = deal_data.take_profit.map(lambda take_profit: float(take_profit))
deal_data["stop_loss"] = deal_data.stop_loss.map(lambda stop_loss: float(stop_loss))
deal_data["order_num"] = deal_data.order_num.map(lambda order_num: int(order_num))
deal_data["commission"] = deal_data.commission.map(lambda commission: float(commission))
deal_data["swap"] = deal_data.swap.map(lambda swap: float(swap.replace(" ", "")))
deal_data["profit"] = deal_data.profit.map(lambda profit: float(profit.replace(" ", "")))
```

```python
deal_data.head(5)
```

```python
# Save DataFrame as csv.
filename = "expert_" + str(magic_num) + "_data.csv"
deal_data.set_index("date_time", inplace = True)
deal_data.to_csv(filename)
```
